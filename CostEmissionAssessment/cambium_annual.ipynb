```python
import pandas as pd 
import numpy as np
df = pd.read_csv(r'G:\University of Michigan Dropbox\Jiahui Chen\Ford_CC\Cambium\\StdScen21_MidCase_annual_balancingArea.csv', skiprows=4)
```

```python
years = ['2024','2030','2040']
balst = ['p'+str(i+1) for i in range(134)]
fuels = ['coal_MWh','gas-cc_MWh','hydro_MWh','wind-ons_MWh','wind-ofs_MWh','csp_MWh','upv_MWh','distpv_MWh','nuclear_MWh']
df_fnal = pd.DataFrame([np.zeros(134) for i in range(len(fuels)*3)])
df_fnal = df_fnal.T
df_fnal.columns = [['2024','2030','2040'][yi] + fuels[fi] for yi in range(3) for fi in range(9)]
for yi in range(3):
    df_fnal[['2024','2030','2040'][yi] + 'solar'] = np.zeros(134)
    df_fnal[['2024','2030','2040'][yi] + 'wind'] = np.zeros(134)
    df_fnal[['2024','2030','2040'][yi] + 'nuchyd'] = np.zeros(134)
df_fnal['bai'] = np.zeros(134)

for bai in range(134):
    for yeari in range(3):
        df_fnal['bai'][bai]=bai
        df1 = df.loc[df['r']==balst[bai]][df['t']==int(years[yeari])]
        for fueli in range(len(fuels)):
            df_fnal[years[yeari]+fuels[fueli]][bai] = df1[fuels[fueli]]/df1['busbar_load'].to_list()[0]
for yi in range(3):
    for fi in [3,4]:
        df_fnal[['2024','2030','2040'][yi] + 'wind'] = df_fnal[['2024','2030','2040'][yi] + 'wind'] + df_fnal[['2024','2030','2040'][yi] + fuels[fi]]
for yi in range(3):
    for fi in [5,6,7]:
        df_fnal[['2024','2030','2040'][yi] + 'solar'] = df_fnal[['2024','2030','2040'][yi] + 'solar'] + df_fnal[['2024','2030','2040'][yi] + fuels[fi]]
for yi in range(3):
    for fi in [2,8]:
        df_fnal[['2024','2030','2040'][yi] + 'nuchyd'] = df_fnal[['2024','2030','2040'][yi] + 'nuchyd'] + df_fnal[['2024','2030','2040'][yi] + fuels[fi]]
```

```python
df_fnal.to_csv(r'G:\Dropbox (University of Michigan)\Ford_CC\Cambium\\annual.csv')
```
