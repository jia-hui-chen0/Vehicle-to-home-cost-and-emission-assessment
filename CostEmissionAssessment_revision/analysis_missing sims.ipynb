```python
import os
import pandas as pd
# Get list of all expected county numbers
expected_counties = [f'{i}_0_0' for i in range(432)]

# Initialize dictionary to store missing counties for each year
missing_counties = {}

# Check each year
for year in ['2024', '2030', '2040']:
    # Get list of existing directories
    try:
        existing_dirs = os.listdir(rf'C:\Users\jiahuic\University of Michigan Dropbox\Jiahui Chen\Ford_CC\Result_v2h\Results_80_Y82_{year}_cons_nocarb')
        # Convert to integers and filter for valid county numbers
        existing_counties = [d for d in existing_dirs if d[:-4].isdigit()]
        # Find missing counties
        missing = [c for c in expected_counties if c not in existing_counties]
        missing_counties[year] = missing
        print(f"Year {year}: {len(missing)} missing counties")
        print(f"Missing counties for {year}: {missing}")
    except FileNotFoundError:
        print(f"Directory for year {year} not found")
        missing_counties[year] = expected_counties

# Create DataFrame of missing counties
missing_df = pd.DataFrame({
    '2024': pd.Series(missing_counties['2024']),
    '2030': pd.Series(missing_counties['2030']),
    '2040': pd.Series(missing_counties['2040'])
})

# Save to CSV
# missing_df.to_csv('missing_counties.csv', index=False)
missing_df
```
