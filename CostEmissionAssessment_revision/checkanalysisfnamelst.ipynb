```python
import os
outputdir = r'C:\Users\jiahuic\Dropbox (University of Michigan)\Ford_CC\\'

# Get the first county's file list as reference
first_county_fnames = os.listdir(outputdir+r'\\Energy consumption_adjusted_1_2_Y60_db\\0')
first_county_fnames = [i for i in first_county_fnames if i[:6]=='0_0_0_']
first_county_fnames = [i for i in first_county_fnames if r'.csv' in i][:15]

# Compare other counties against the first county
for county_num in range(1, 432):
    try:
        current_fnames = os.listdir(outputdir+r'\\Energy consumption_adjusted_1_2_Y60_db\\'+str(county_num))
        current_fnames = [i for i in current_fnames if i[:6]=='0_0_0_']
        current_fnames = [i for i in current_fnames if r'.csv' in i][:15]
        
        if current_fnames != first_county_fnames:
            print(f"County {county_num} has different files:")
            print(f"Expected: {first_county_fnames}")
            print(f"Found: {current_fnames}")
            print("---")
    except FileNotFoundError:
        print(f"County {county_num} directory not found")
```
